```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import os
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
%matplotlib inline
```

```python
post_dir = "../_posts/"
img_dir = "../assets/clouds/"
allposts = os.listdir(post_dir)
```

```python
fp = open(post_dir + allposts[0],'r')
text = fp.read()
wordcloud = WordCloud(max_words=30,background_color=(21,21,21)).generate(text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
for fn in allposts:
    if fn.endswith(".md") and os.path.isfile(fn):
        fp = open( post_dir + fn,'r')
        text = fp.read()
        wordcloud = WordCloud(max_words=30,background_color=(21,21,21)).generate(text)
        outname = img_dir + fn[:-len('md')] + 'png'
        wordcloud.to_file(outname)
```
