```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import os
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
%matplotlib inline
```

```python
allposts = os.listdir("_posts")
allposts
```

```python
fp = open("_posts/" + allposts[0],'r')
text = fp.read()
wordcloud = WordCloud(max_words=30,background_color=(21,21,21)).generate(text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python

for fn in allposts:
    if fn.endswith(".md"):
        fp = open( "_posts/" + fn,'r')
        text = fp.read()
        wordcloud = WordCloud(max_words=30,background_color=(21,21,21)).generate(text)
        outname = "assets/clouds/" + fn[:-len('md')] + 'png'
        wordcloud.to_file(outname)
```
